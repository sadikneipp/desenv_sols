```python
import pandas as pd
from glob import glob
```

```python
common_keys = ['SUBJECT_ID', 'HADM_ID']

df = pd.DataFrame(columns=common_keys)
later = ['./data/CHARTEVENTS.csv', './data/INPUTEVENTS_MV.csv', './data/LABEVENTS.csv',
         './data/PRESCRIPTIONS.csv', './data/DATETIMEEVENTS.csv', './data/PROCEDURES_ICD.csv']
for path in glob('./data/*'):
    if path in later: continue
    try:
        new = pd.read_csv(path)
        if len(new) > 1000: continue
        print(path)
        df = pd.merge(df, new, on=common_keys, how='outer') 
    except KeyError:
        print('key not found in' + path)
        continue      
df
```

```python
df
```

```python
def fillna_frequent(x):
    try:
        return x.fillna(x.value_counts().index[0])
    except IndexError:
        return x
```

```python
new_df = pd.DataFrame(columns=df.columns)
for tup, subdf in df.groupby(by=['HADM_ID']):
    subdf = subdf.apply(fillna_frequent)
    new_df = pd.concat([subdf, new_df])
    
new_df = new_df.reset_index(drop=True)
new_df
```

```python
nan_dict ={}
for column in list(new_df.columns):
    print('New: ' + str(new_df[column].isnull().sum()) )
    print('Old: ' + str(df[column].isnull().sum()) )
```
